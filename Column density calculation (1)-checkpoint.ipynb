```python
import numpy as np
import matplotlib.pyplot as plt
import math

molecule = input("Hello. What molecule do you want to measure?: ").upper()
if molecule == 'CO':
    B = 5.7635968E10 #in Hz
    mu = 1.1011E-19 #in esu
elif molecule == 'HCN':
    B = 4.4315976E10 
    mu = 2.9852E-18
elif molecule == 'HCO+':
    B = 4.459443E10
    mu = 3.90E-18
else: 
    print("I'm sorry, that molecule isn't in our database.")

beam_size = []
frequency = []
J = []
beam_efficiency = []
transitionlist = []
transition = ''

count = 0
while transition != 'done':
    transition = input("What transition? Enter 'done' when finished: ")
    if molecule == 'CO' and transition == '1-0':
        frequency.append(115271201.8) #in Hz
        J.append(1) #upper energy level J
        beam_size.append(54.5763590712) #in arcsec
        beam_efficiency.append(0.84)
        transitionlist.append(transition)
        count += 1
    elif molecule == 'CO' and transition == '2-1':
        frequency.append(230538000)
        J.append(2)
        beam_size.append(32.7464409338)
        beam_efficiency.append(0.75)
        transitionlist.append(transition)
        count += 1
    elif molecule == 'CO' and transition == '3-2':
        frequency.append(345795989.9)
        J.append(3)
        beam_size.append(21.8316557175)
        beam_efficiency.append(0.66)
        transitionlist.append(transition)
        count += 1
    elif molecule == 'HCN' and transition == '1-0':
        frequency.append(88631602.2)
        J.append(1)
        beam_size.append(70.9801283498)
        beam_efficiency.append(0.88)
        transitionlist.append(transition)
        count += 1
    elif molecule == 'HCN' and transition == '3-2':
        frequency.append(265886433.9)
        J.append(3)
        beam_size.append(28.3929453988)
        beam_efficiency.append(0.75)
        transitionlist.append(transition)
        count += 1
    elif molecule == 'HCO+' and transition == '1-0':
        frequency.append(89188524.7)
        J.append(1)
        beam_size.append(70.5369050689)
        beam_efficiency.append(0.88)
        transitionlist.append(transition)
        count += 1
    elif molecule == 'HCO+' and transition == '3-2':
        frequency.append(267557625.9)
        J.append(3)
        beam_size.append(28.2156001893)
        beam_efficiency.append(0.75)
        transitionlist.append(transition)
        count += 1
    elif transition != 'done':
        print("I'm sorry, that transition isn't in our database.")
        


source_size = input("Next, what is the source size in arcseconds?: ")

temperatures = []

for i in range(count):
    peak_temp = input("What is the peak temperature of the " + molecule + " " + transitionlist[i] + " line in units of K?: ")
    if peak_temp != 'done':
        peak_temp = float(peak_temp)
        temperatures.append(peak_temp)
        
#check to make sure count = number of temperatures entered

print('\nHere is your list of transitions and corresponding intensities:\n')
for i in range(len(transitionlist)):
    print(molecule + " " + transitionlist[i] + ": " + str(temperatures[i]))

widths = []

for i in range(count):
    line_width = input("Lastly, what is the line width of the " + molecule + " " + transitionlist[i] + " line?: ")
    if line_width != 'done':
        line_width = float(line_width)
        widths.append(line_width)

k = 8.617333E-5 #ev/K
h = 4.135667696E-15 #ev/Hz

#a=T_r
numerator = []
denominator = []
#c = numerator/denominator
xs = []
ys = []
for i in range(len(temperatures)):
    math1 = (float(temperatures[i]))/((float(source_size)**2))
    math2 = ((float(source_size)**2) + (float(beam_size[i])**2))
    math3 = (1/(float(beam_efficiency[i])))
    a = math1*math2*math3
    numerator.append(3*k*a*widths[i])
    denominator.append(248.0502134*frequency[i]*J[i]*(mu**2))
    c = (numerator[i])/(denominator[i])
    ys.append(math.log(c, 10))
    math4 = h*B*float(J[i])
    math5 = float(J[i]) + 1
    xs.append((math4*math5)/k)

xfloats = []
yfloats = []
for item in xs:
    xfloats.append(float(item))
for item in ys:
    yfloats.append(float(item))
fig = plt.figure(figsize = (6, 3))
m, b = np.polyfit(xfloats, yfloats, 1)
plt.scatter(xs, ys, s=100)
x_line = np.arange(min(xfloats), (max(xfloats)+ 1), 1)
y_line = m*x_line + b
plt.plot(x_line, y_line, '--', color='red')
plt.show()
mstring = str(m)
bstring = str(b)
print("This is the slope of the line of best fit: " + mstring)
print("And this is the y-intercept: " + bstring)

    
T_rot = (-0.4342944819)/(m)
N_tot = ((pow(10, b))*(T_rot))/(B)

print("This is your column density: " + "{:e}".format(N_tot))

  
```
