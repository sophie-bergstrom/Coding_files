```python
import numpy as np
import matplotlib.pyplot as plt

#Order of PNe for HCN
#Hb 5, K3-58, K3-45, NGC6772, K4-47, M1-7, M3-28, NGC2440, M3-55, M4-14, K3-17, He2-459, Hu1-1, K3-52, K3-65, M1-59, M1-8, M2-53, M4-17, NGC6445

#Order of PNe for HCO+
#Hb 5, K3-58, K4-47, M1-7, M3-28, NGC2440, M2-9, M3-55, M4-14, K3-83, K3-17, He2-459, Hu1-1, K3-52, K3-65, M1-59, M1-8, M2-53, M4-17, NGC6445

#need to change abundances by doing CO column density/H2 density = 10^-4 and find H2 density, then do HCN/HCO+ density/H2 density
#experimentally determined
time_HCN = np.asarray([1500, 11070, 830, 10900, 900, 6000, 7700, 2960, 5040, 5550, 970, 1900, 7960, 4000, 7000, 4700, 8010, 20400, 6100, 14900]) #time/age of nebula
abundance_HCN = np.asarray([9.1E-7, 2.3E-7, 1.8E-8, 9.2E-8, 7.6E-7, 1.4E-8, 9.8E-9, 1.7E-7, 4.2E-8, 1.1E-7, 6.9E-8, 2.0E-8, hu11, 4.0E-6, 1.8E-8, 3.5E-8, m18, m253, 4.3e-8, 9.0E-8]) #abundance of the molecule in each nebula

time_HCO = np.asarray([1500, 11070, 900, 6000, 7700, 2960, 2500, 5040, 5550, 4600, 970, 1900, 7960, 4000, 7000, 4700, 8010, 20400, 6100, 14900])
abundance_HCO = np.asarray([7.4E-7, 1.4E-7, 4.5E-8, 4.7E-9, 3.5E-9, 3.8E-8, 9.7E-8, 6.8E-9, 3.1E-8, 5.8E-9, 2.2E-8, 1.5E-6, hu11, 8.8E-7, 5.2E-9, 1.8E-8, m18, m253, 1.5e-8, 2.1E-8])

#model from Redman et al. (2003)
time_model_HCN = np.asarray([2550, 6300, 10050]) #time/age of nebula
abundance_model_HCN = np.asarray([8.0E-5, 2.0E-10, 2.6E-12]) #modeled abundance of molecule
time_model_HCO = np.asarray([2550, 6300, 10050])
abundance_model_HCO = np.asarray([2.2E-10, 8.8E-12, 1.2E-11])

#time_HCN2 = np.asarray([]) #for NGC7027, NGC6537, M248, NGC6720, and NGC7293
#abundance_HCN2 = np.asarray([])
#time_HCO2 = np.asarray([]) #for NGC7027, K447, NGC6537, M248, NGC6720, NGC6853, and NGC7293
#abundance_HCO2 = np.asarray([])

abundance_error_HCN = np.asarray([3.5E-7, 1.0E-7, 0.8E-8, 3.3E-8, 2.6E-7, 0.8E-8, 5.5E-9, 0.8E-7, 2.0E-8, 0.5E-7, 3.5E-8])
abundance_error_HCO = np.asarray([2.9E-7, 0.6E-7, 0, 2.7E-9, 1.9E-9, 1.8E-8, 4.3E-8, 3.1E-9, 1.4E-8, 3.3E-9, 1.0E-8])

fig=plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.set_yscale('log')

plt.scatter(x=time_HCN, y=abundance_HCN, color="r")
plt.scatter(x=time_HCO, y=abundance_HCO, color="b")
plt.plot(time_model_HCN, abundance_model_HCN, marker="^", color="green", linestyle="--")
plt.plot(time_model_HCO, abundance_model_HCO, marker="^", color="purple", linestyle="--")
#plt.scatter(x=time_HCN2, y=abundance_HCN2, color="r", marker="*")
#plt.scatter(x=time_HCO2, y=abundance_HCO2, color="b", marker="*")
plt.errorbar(time_HCN, abundance_HCN, yerr=abundance_error_HCN, fmt="o", color="r")
plt.errorbar(time_HCO, abundance_HCO, yerr=abundance_error_HCO, fmt="o", color="b")
plt.xlabel("Time(years)")
plt.ylabel("Abundance(units)")
plt.title("Abundance vs Time")
plt.legend(["Experimentally Determined", "Modeled"])

plt.show()

```
